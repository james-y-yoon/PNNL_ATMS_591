```python
import pandas as pd
import numpy as np
import xarray as xr
import glob as glob
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('ticks')
%config InlineBackend.figure_format='retina'
```

# PDFs of colors

```python
red = np.loadtxt('Image_data_R.dat', delimiter = ',')
green = np.loadtxt('Image_data_G.dat', delimiter = ',')
blue = np.loadtxt('Image_data_B.dat', delimiter = ',')
panchromatic = np.loadtxt('Image_data_panchromatic.dat', delimiter = ',')
```

```python
red_flatten = red.flatten()
green_flatten = green.flatten()
blue_flatten = blue.flatten()
panchromatic_flatten = panchromatic.flatten()

fig, (ax1, ax2, ax3, ax4) = plt.subplots(4,1, sharex = True, figsize = (7,10))

ax1.hist(red_flatten, bins = 60, color = 'tab:red')
ax2.hist(green_flatten, bins = 60, color = 'tab:green')
ax3.hist(blue_flatten, bins = 60, color = 'tab:blue')
ax4.hist(panchromatic_flatten, bins = 60, color = 'tab:orange')

ax1.set_title('Red')
ax2.set_title('Green')
ax3.set_title('Blue')
ax4.set_title('Panchromatic')

fig.suptitle('Color indices in BB_Smoke_Over_Tricities', weight = 'bold', y=0.93)

ax4.set_xlabel('Color index')
ax1.set_ylabel('Count')
ax2.set_ylabel('Count')
ax3.set_ylabel('Count')
ax4.set_ylabel('Count')
```

```python
fig, ax = plt.subplots(2,2, figsize = (7,5))

im = ax[0,0].imshow(red, cmap = 'Reds')
plt.colorbar(im)

im = ax[1,0].imshow(green, cmap = 'Greens')
plt.colorbar(im)

im = ax[0,1].imshow(blue, cmap = 'Blues')
plt.colorbar(im)

im = ax[1,1].imshow(panchromatic, cmap = 'Greys_r')
plt.colorbar(im)

# plt.setp(plt.gcf().get_axes(), xticks=[], yticks=[]);
ax[0,0].set_title('Red')
ax[1,0].set_title('Greens')
ax[0,1].set_title('Blues')
ax[1,1].set_title('Panchromatic')
```

```python
fig, ax = plt.subplots(2,2, figsize = (7,5))

im = ax[0,0].imshow(red / green, cmap = 'Reds')
plt.colorbar(im)

im = ax[1,0].imshow(blue / red, cmap = 'Blues')
plt.colorbar(im)

im = ax[0,1].imshow(blue / green, cmap = 'Blues')
plt.colorbar(im)

im = ax[1,1].imshow(panchromatic, cmap = 'Greys_r')
plt.colorbar(im)

# plt.setp(plt.gcf().get_axes(), xticks=[], yticks=[]);
ax[0,0].set_title('Red / Green')
ax[1,0].set_title('Blue / Red')
ax[0,1].set_title('Blue / Green')
ax[1,1].set_title('Panchromatic')
```
