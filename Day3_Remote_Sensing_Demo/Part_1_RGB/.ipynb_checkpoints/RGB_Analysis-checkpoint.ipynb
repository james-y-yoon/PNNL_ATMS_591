# Part 1: RGB Analysis

A true color image (file name: “Image_BBSmoke_Over_Tricity.jpg”) was taken over Richland WA during biomass burning event. The true color image can be saved in limited specific formats (e.g., jpg, png, tiff etc). These formats write image data in three-dimensional data matrix (e.g., 1613 x 2169 x 3 for the give image). 

The first two dimensions are height and width of the image (number of pixels). The third dimension is Red (R), Green (G), and Blue (B) color indexes (0-255) for the given pixels.

- Open the True color image file: Image_BBSmoke_Over_Tricity.jpg
- Open the same image into panchromatic format (Image_BBSmoke_Over_Tricity_panchromatic.jpg)
- Have a look on the individual R, G, B color indexes/components of the data from true color image.


```python
import pandas as pd
import numpy as np
import xarray as xr
import glob as glob
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('ticks')
%config InlineBackend.figure_format='retina'
```

```python
red = np.loadtxt('Image_data_R.dat', delimiter = ',')
green = np.loadtxt('Image_data_G.dat', delimiter = ',')
blue = np.loadtxt('Image_data_B.dat', delimiter = ',')
panchromatic = np.loadtxt('Image_data_panchromatic.dat', delimiter = ',')
```

# PDFs of colors

- Try to investigate the pdf distribution of these RGB components
- Try to investigate the pdf distribution of the panchromatic image
- Do the pdfs from (d) and (e) look same or different? Why?

```python
red_flatten = red.flatten()
green_flatten = green.flatten()
blue_flatten = blue.flatten()
panchromatic_flatten = panchromatic.flatten()

fig, (ax1, ax2, ax3, ax4) = plt.subplots(4,1, sharex = True, figsize = (7,10))

ax1.hist(red_flatten, bins = 60, color = 'tab:red')
ax2.hist(green_flatten, bins = 60, color = 'tab:green')
ax3.hist(blue_flatten, bins = 60, color = 'tab:blue')
ax4.hist(panchromatic_flatten, bins = 60, color = 'tab:orange')

ax1.set_title('Red')
ax2.set_title('Green')
ax3.set_title('Blue')
ax4.set_title('Panchromatic')

ax1.set_xlim([0, 255])
fig.suptitle('Color indices in BB_Smoke_Over_Tricities', weight = 'bold', y=0.93)

ax4.set_xlabel('Color index')
ax1.set_ylabel('Count')
ax2.set_ylabel('Count')
ax3.set_ylabel('Count')
ax4.set_ylabel('Count')
```

## Plotting image channels and ratios

```python
fig, ax = plt.subplots(2,2, figsize = (7,5))

im = ax[0,0].imshow(red, cmap = 'Reds')
cb = plt.colorbar(im)

im = ax[1,0].imshow(green, cmap = 'Greens')
cb = plt.colorbar(im)

im = ax[0,1].imshow(blue, cmap = 'Blues')
cb = plt.colorbar(im)

im = ax[1,1].imshow(panchromatic, cmap = 'Greys_r')
cb = plt.colorbar(im)

for ax_i in ax.ravel():
    ax_i.set_xticks([])
    ax_i.set_yticks([])

ax[0,0].set_title('Red')
ax[1,0].set_title('Green')
ax[0,1].set_title('Blue')
ax[1,1].set_title('Panchromatic')
```

```python
fig, ax = plt.subplots(2,2, figsize = (7,5))

im = ax[0,0].imshow(red / green, cmap = 'Reds', vmax = 3)
plt.colorbar(im)

im = ax[1,0].imshow(blue / red, cmap = 'Blues', vmax = 3)
plt.colorbar(im)

im = ax[0,1].imshow(green / blue, cmap = 'Greens', vmax = 3)
plt.colorbar(im)

im = ax[1,1].imshow(panchromatic, cmap = 'Greys_r')
plt.colorbar(im)

# plt.setp(plt.gcf().get_axes(), xticks=[], yticks=[]);
ax[0,0].set_title('Red / Green')
ax[1,0].set_title('Blue / Red')
ax[0,1].set_title('Green / Blue')
ax[1,1].set_title('Panchromatic')

for ax_i in ax.ravel():
    ax_i.set_xticks([])
    ax_i.set_yticks([])
```
